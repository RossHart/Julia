```julia
# we can define a function quite easily in Julia:
function square_my_number(number)
    return number ^ 2
end

println("Result of square_my_number(2) = $(square_my_number(2))")
 
# Multiple dispatch -- we can make our function act 
# differently dependent on what data we give it like this
function cube_my_number(number::Number)
    return number ^ 3
end

function cube_my_number(number::String)
    return "I cannot cube a string!"
end

# print the results
println("Result of cube_my_number(2) = $(cube_my_number(2))")
println("""Result of cube_my_number("2") = $(cube_my_number("2"))""")
# we can also check which method we used to get an answer
println(@which cube_my_number(2))
```

```julia
# we can also define anonymous functions 
# (like Python's lambda functions)
x -> x^2
# is equivalent to:
function square_my_number(number)
    return number ^ 2
end

# We can apply these functions to an array using broadcast
my_array = [1, 2, 3]
my_squared_array = broadcast(x -> x^2, my_array)
my_other_squared_array = broadcast(square_my_number, my_array)

# print the results
println("my_squared_array = $my_squared_array")
println("my_other_squared_array = $my_other_squared_array")
```

```julia
# Julia base functions have two types
my_array_1 = [5, 3, 2, 1, 4]
my_array_sorted_1 = sort(my_array_1)

# Using ! means operations happen in-place
my_array_2 = [1, 2, 3, 4, 5]
my_array_sorted_2 = sort!(my_array_2);
```

```julia
println("Using sort:")
println("my_array_1 = ", my_array_1)
println("my_array_sorted_1 = ", my_array_sorted_1)

println("\nUsing sort!:")
println("my_array_2 = ", my_array_2)
println("my_array_sorted_2 = ", my_array_sorted_2)
```
